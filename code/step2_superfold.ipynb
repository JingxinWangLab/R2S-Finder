```python
# Run batch process from here! 
# Python 2 is used here
import os
import subprocess
from concurrent.futures import ThreadPoolExecutor
import shutil
```

```python
# human_invivo
def run_script(file_name):
    """Function to run a script on a file."""
    command = "python Superfold.py {}".format(file_name)
    try:
        result = subprocess.call(command, shell=True)
        if result == 0:
            return "Successfully processed {}".format(file_name)
        else:
            return "Failed to process {}".format(file_name)
    except Exception as e:
        return "Error processing {}: {}".format(file_name, str(e))

def main():
    # List all .map files in the current directory
    files = [f for f in os.listdir('.') if f.endswith('.map')]
    
    # Using ThreadPoolExecutor to manage multiple threads
    with ThreadPoolExecutor(max_workers=12) as executor:
        # Map the run_script function to each file
        results = list(executor.map(run_script, files))
    
    # Print results
    for result in results:
        print(result)

if __name__ == '__main__':
    main()
```

```python
temp_map_result_dir = './human_invivo_map_result'
temp_partition_fold_dir = './human_invivo_partition_fold'

if not os.path.exists(temp_map_result_dir):
    os.makedirs(temp_map_result_dir)

if not os.path.exists(temp_partition_fold_dir):
    os.makedirs(temp_partition_fold_dir)

items = os.listdir('./')
for item in items:
    # Move .map files to ./temp_map_result
    if item.endswith('.map') or (os.path.isdir(item) and item.startswith('results_')):
        shutil.move(item, temp_map_result_dir)
    # Move folders starting with "partition_" or "fold_" to ./temp_partition_fold
    elif os.path.isdir(item) and (item.startswith('partition_') or item.startswith('fold_')):
        shutil.move(item, temp_partition_fold_dir)
```
